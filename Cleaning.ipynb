```python
import pandas as pd
```

```python
file="KPMG_VI_New_raw_data_update_final (1).xlsx"
transDf=pd.read_excel(file,sheet_name="Transactions")
transDf.head()
```

```python
transDf.isna().sum()
```

```python
transDf.describe()
```

```python
transDf.dropna(axis=0,inplace=True)
```

```python
transDf.info()
```

```python
# transDf.set_index("transaction_id",inplace=True)
transDf.tail()
```

```python
transDf.columns
```

```python
transDf.online_order.value_counts()
```

```python
transDf[ transDf["online_order"]!=0].head()
```

```python
transDf.online_order.replace({True:1,False:0},inplace=True)
transDf
```

```python
print(transDf.product_size.value_counts())
transDf.product_size.replace({"medium":2,"small":1,"large":3},inplace=True)
transDf.head()
```

```python
print(transDf.product_class.value_counts())
transDf.product_class.replace({"medium":2,"low":1,"high":3},inplace=True)
transDf.head()
```

```python
print(transDf.order_status.value_counts())
order_status_dummies=pd.DataFrame(pd.get_dummies(transDf.order_status))
transDf["Approved_Orders"]=order_status_dummies.Approved
transDf["Cancelled_Orders"]=order_status_dummies.Cancelled
transDf.head()
```

```python
transDf.corr()
```

```python
customer_grp=transDf[['transaction_id', 'product_id', 'customer_id', 'transaction_date','order_status', 
                      'product_class', 'product_size', 'list_price', 'standard_cost', 'Approved_Orders', 'Cancelled_Orders']].groupby(["customer_id"]).agg(
                          {
                              
           'transaction_id': "count", 
           'product_id':"median", 
           'transaction_date': "max",
           'product_class':"mean", 
           'product_size':"max", 
           'list_price':["max","mean","min"], 
           'standard_cost':["max","mean","min"], 
           'Approved_Orders':"sum", 
           'Cancelled_Orders':"sum"
           })
print(customer_grp.info())
customer_grp.head()
```

```python
import seaborn as sns
sns.heatmap(customer_grp.corr(),robust=True,color="red")
```
