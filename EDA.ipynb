```python
df_grouped=df.groupby("Age").agg({"product_id":"max",
                                          "Age":"mean",
                                          "transaction_date":"min",
                                          "list_price":"sum",
                                          "standard_cost":"sum"})
df_grouped.head()
```

```python
fig = px.bar(x=df_grouped.Age, y = df_grouped.list_price)
fig.update_xaxes(type='category')
fig.show()
```
